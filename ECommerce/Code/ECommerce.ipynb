## create RMF model by each customer

```python
# create RMF model by each customer

RMFScore = df.groupby('CustomerID').agg({'Date1': lambda x:(latest_date-x.max()).days,
                                            'Date': lambda x:x.count(),
                                            'Price': lambda x:x.sum()})
```

```python
RMFScore['Date'] = RMFScore['Date'].astype(int)
RMFScore.rename(columns={'Date1':'Recency','Date':'Frequency','Price':'Monetory'},inplace=True)


RMFScore.reset_index().head()
```

```python
RMFScore.Recency.describe()
```

## Split data into four segments using quantile method

```python
quantiles = RMFScore.quantile(q=[0.25,0.50,0.75])
quantiles = quantiles.to_dict()
```

```python
quantiles
```

```python
## So for recency , lower the recency better the customer
## for frequency , maximum the freq better the cust
## for monetory , higher the sum of total amount that is my better customer
```

### Create own function to define R, F and M segment

```python
def RScoring(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4
    
def FnMScoring(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]:
        return 2
    else:
        return 1
```

```python
RMFScore.columns
```

```python
RMFScore['R']= RMFScore['Recency'].apply(RScoring,args=('Recency',quantiles,))
RMFScore['F']= RMFScore['Frequency'].apply(RScoring,args=('Frequency',quantiles,))
RMFScore['M']= RMFScore['Monetory'].apply(RScoring,args=('Monetory',quantiles,))
```

```python
RMFScore
```

```python
RMFScore['RMFGroup']= RMFScore['R'].map(str) + RMFScore['F'].map(str) + RMFScore['M'].map(str)
RMFScore.head()
```

```python
RMFScore.RMFGroup=RMFScore['RMFGroup'].astype(int)
```

```python
loyalty_level=['Diamond','Platinum','Gold','Silver']
score_cuts = pd.qcut(RMFScore.RMFGroup,q=4,labels=loyalty_level)
RMFScore['RMFScore']=score_cuts.values
RMFScore.reset_index().head(20)
```

# Build Clustering Model

```python
from sklearn.cluster import KMeans
```

```python
sum_of_sq_dist = {}
for k in range(1,15):
    km = KMeans(n_clusters=k,init = 'k-means++',max_iter = 300)
    km = km.fit(scaled_data)
    sum_of_sq_dist[k] = km.inertia_
    
```

```python
#Visualization for getting elbow method (to find the actual K value)
sns.pointplot(x=list(sum_of_sq_dist.keys()),y=list(sum_of_sq_dist.values()))
plt.xlabel("Number of Clusters(k) ")
plt.ylabel("Sum of square distance - eucledian")
plt.show()
```

```python
km = KMeans(n_clusters=4,init='k-means++',max_iter=300)
y_means=km.fit_predict(scaled_data)
```

```python
y_means
```

```python
RMFScore['Cluster']=km.labels_
RMFScore.head()
```
