```python
import ast 
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
```

```python
movies_df.head(2)
```

### Extract Keyword names

```python
movies_df['keywords'] = movies_df['keywords'].apply(convert)
```

### Extract top 3 cast members from cast 

```python
def convert_cast(text):
    L=[]
    counter = 0
    for i in ast.literal_eval(text):
        if counter<3:
            L.append(i['name'])
        counter=counter+1
    return L
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert_cast)
```

```python
movies_df.head(2)
```

### Extract Director from crew

```python
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=="Director":
            L.append(i['name'])
    return L
```

```python
movies_df['crew'] = movies_df['crew'].apply(fetch_director)
```

```python
movies_df.head(2)
```

#### Now remember we want movie_id , title ,runtime,vote_average tags (genres,overview,cast,crew)

##### But at the same time , look closely at cast and crew there could be multiple people with same first name and that 
##### can confuse the machine 
##### So we need to append the first and last name or any text column & remove spaces 
#### Otherwise machine will consider first and last name or anythig with spaces as two unique vectors or number

```python
def removespaces(text):
    L1=[]
    for i in text:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies_df['genres'] = movies_df['genres'].apply(removespaces)
movies_df['keywords'] = movies_df['keywords'].apply(removespaces)
movies_df['cast'] = movies_df['cast'].apply(removespaces)
movies_df['crew'] = movies_df['crew'].apply(removespaces)
```

### stopword removal

```python
import nltk
```

```python

# nltk.download()
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
def remove_stopwords(obj):
    final_sentence = []
    tokens = word_tokenize(obj)
    for each_word in tokens:
        if each_word not in stopwords.words('english'):
            final_sentence.append(each_word)
            
    return " ".join(final_sentence)


            
```

```python
movies_final['tags'] = movies_final['tags'].apply(remove_stopwords) 
```

### Stemming 

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stemming(text):
    result = []
    for i in text.split():
        result.append(ps.stem(i))
    return " ".join(result)
```

```python
movies_final['tags'] = movies_final['tags'].apply(stemming)
```

### Recommendation logic

```python
def find_closest(text):
    spliter = movies_final['title'].str.split(' ')
    i=0
    for val in spliter:
        if text in val:
            print(movies_final['title'][i])
            return i
        i+=1
```

```python
def recommend_with_bert(movie):
    movie = movie.lower()
    titles = movies_final['title']
    if movie in titles.to_list():
        index = movies_final[movies_final['title']==movie].index[0]
    else:
        index = find_closest(movie)
        
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:10]:
        print(movies_final.iloc[i[0]].title)
```
